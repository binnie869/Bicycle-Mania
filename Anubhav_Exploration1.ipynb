### Fixing End time

```python
for i in range(len(df)):
    start_dt = datetime.strptime(df['Start Date'].values[i], '%m/%d/%Y %H:%M')
    end_dt = datetime.strptime(df['End Date'].values[i], '%m/%d/%Y %H:%M')
    delta = timedelta(minutes=df['Duration'].values[i])
    actual_end_dt = end_dt+delta
    df['End Date'][i] = actual_end_dt.strftime('%m/%d/%Y %H:%M')
```
